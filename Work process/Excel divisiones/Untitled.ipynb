```python
import pandas as pd
import numpy as np

archivoOrigen='Generación Intermitente de la GCROC 2013-2020.xlsx'

extension='.xlsx'
nombreHoja='Datos'
ArchivoDestino=[]

listaHoras=[]
lista=[]
Hora='hora'
RenTitulos=3 # Numero de Renglón +1 en donde se encuentran los títulos
ColTitulos=4 # Numero de Columna -1 en donde se encuentran los títulos


```

```python
#GICIE301
# Proceso para Generar nuevo Archivo en Excel de 2 filas 
datos=pd.read_excel(archivoOrigen, header=1)

# Extrae los datos de los Títulos para
# para despues usarla para generar los archivos .xlxs
ArchivoDestino=datos.iloc[[RenTitulos],]


# Genera una Lista para Encabezados de Hora
for lH in range(1,25): 
    listaHoras.append("Hora "+str(lH))
```

```python

#print(listaHoras)
#print(datos[Hora])


cont=0


for AD in ArchivoDestino:
    if cont>=ColTitulos:
        Titulos=AD+extension
        for ciclos in datos[AD]:
            lista.append(ciclos)
        arregloCiclos=np.array(lista)
        i=0

        #print(listaCiclos[0:24])
        df=pd.DataFrame()
        for hC in range(i,i+24):
            df.loc[:,listaHoras[i]]=arregloCiclos[hC:hC+1]
            i+=1
        
        h=24
        
        lon=(int(len(arregloCiclos)/24))
        #print(type(lon))
        #print("longitud:", lon)
        for i in range(1,lon):
            df.loc[len(df)]=arregloCiclos[h:h+24]
            h=h+24
            
        #df.loc[len(df)]=listaCiclos[24:48]  
        #df.loc[len(df)]=listaCiclos[48:72]  
        #print(len(listaCiclos))

        #df.loc[:,listaHoras[0]]=listaCiclos[0:1]
        #df.loc[:,listaHoras[1]]=listaCiclos[1:2]
        
        #df.loc[:,listaHoras[23]]=listaCiclos[23:24]
        
        #print(listaHoras[0])
        #print(listaCiclos[0])
        
        #df.loc=listaCiclos[0:24]
        #print(datos[AD])
        print("Nombre Archivo: ", Titulos)
        grabar = pd.ExcelWriter(Titulos)
        df.to_excel(grabar,nombreHoja)
        grabar.save()
        
        for i in range(len(lista)-1, -1, -1):
            lista.pop()     

        
    cont+=1
```
